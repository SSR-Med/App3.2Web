```python
dataset = pd.read_csv("Recursos/Usuarios.csv")
```

```python
# Seleccionamos la fila del usuario
usuario = "Usuario1"
rowUsuario = dataset.loc[dataset["Usuario"] == usuario]
# Borrar la fila del dataset
dataset.drop(dataset[dataset["Usuario"] == usuario].index, inplace=True)
dataset.reset_index(inplace=True)
dataset.drop('index',axis=1,inplace=True)
```

```python
def similitudUsuarios(usuarioRecomendar, usuario):
    distancia = np.array((usuarioRecomendar-usuario)**2,dtype=float)
    return sum(distancia[~np.isnan(distancia)])**0.5
    
```

```python
def similitudMaker(row):
    return similitudUsuarios(rowUsuario.values[0][1:],row.values[1:])
dataset["Similitud"] = dataset.apply(lambda row:similitudMaker(row),axis=1)
```

```python
def weightMatrix(datasetFuncion):
    return datasetFuncion.iloc[:,1:-1].multiply(datasetFuncion["Similitud"],axis="index")
```

```python
datasetWeight = weightMatrix(dataset)
```

```python
datasetWeight
```

```python
wa = np.array(datasetWeight["Eraserhead"],dtype=float)
sum(wa[~np.isnan(wa)])
```

```python
sumWeightRating = {pelicula:sum(np.array(datasetWeight[pelicula],dtype=float)[~np.isnan(np.array(datasetWeight[pelicula],dtype=float))]) for pelicula in datasetWeight.columns}
```

```python
weightSum = {}
for pelicula in datasetWeight.columns:
    weightSum[pelicula] = sum(dataset[~dataset[pelicula].isna()]["Similitud"])
```

```python
weightSum
```

```python
resultados = {pelicula:round(sumWeightRating[pelicula]/weightSum[pelicula],2) for pelicula in sumWeightRating}
resultados
```

```python
import math
peliculasNoCalificadas = []
for pelicula in rowUsuario.columns[1:]:
    if math.isnan(rowUsuario[pelicula].values[0]) :
        peliculasNoCalificadas.append(pelicula)
```

```python
diccionarioPeliculasNoCalificadas = {pelicula:resultados[pelicula] for pelicula in peliculasNoCalificadas}
```

```python
ordenPeliculasNoCalificadas = {pelicula: score for pelicula, score in sorted(diccionarioPeliculasNoCalificadas.items(), key=lambda item: item[1],reverse=True)[:5]}
ordenPeliculasNoCalificadas
```

```python
datasetResultado = pd.DataFrame({"Peliculas":[valor for valor in ordenPeliculasNoCalificadas],"Rating":[ordenPeliculasNoCalificadas[pelicula] for pelicula in ordenPeliculasNoCalificadas]})
```
